```python
from scipy.stats import gamma
import numpy as np
```

```python
#gamma function with shape factor =1 and rate hyperparameter = 95
f = gamma(95)
```

```python
#calculate 95% equitailed credible set
x = f.interval(0.95)
```

```python
#calculate actual lambda
x = np.array(x) /171.148
```

```python
print("95% Equitailed Credible set for λ", x)
```

```python
#calculationg posterior of hypothesis
posterior = f.cdf(0.5*171.148)
```
